```python
# 노래 차트를 볼 수 있는 :: 지니
# 지니 차트 200위까지 크롤링
# url의 변화를 주면서 for문을 돌려보아라, 단 4번에 걸쳐서 크롤링은 아님
# 수집한 데이터를 데이터프레임으로 저장
# 엑셀로 출력(내보내기)
# 과제는 github에 내보내기

import requests
from bs4 import BeautifulSoup

# header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
# table = soup.find('table', {'class' : 'list-wrap'})
# titles = table.findAll('a', {'class':'title'})
# artists = table.findAll('a', {'class':'artist'})

# for i, (t, a) in enumerate(zip(titles, artists)):
#   titles = t.text.strip() # strip은 공백 없애기
#   artists = a.text.strip().split('\n')[0] # \n을 맨 앞에것만 사용하겠다
#   print('{0:3d}위 {1} - {2}'.format(i+1, artists, titles))

page = ['1','2','3','4']
rank = 1
data2 = []

for i in page:
  genie = f"https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230601&hh=21&rtm=Y&pg={i}"

  header = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64'}
  request = requests.get(genie, headers=header)
  soup = BeautifulSoup(request.text)

  tables = soup.find('table', {'class' : 'list-wrap'})
  titles = tables.findAll('a', {'class':'title'})
  artists = tables.findAll('a', {'class' : 'artist'})
  

  for j, (t, a) in enumerate(zip(titles,artists), start=rank):                        # 이 부분은 배꼈습니다. 특히 랭크관련 
    title = t.text.strip()
    artist = a.text.strip()
    print('{0:3d}위 {1} - {2}'.format(j, title, artist))
    dic = {'ranking':rank, 'titles': title, 'artists': artist} # 배낀부분 2
    data2.append(dic)  #배낀 부분 3
    rank = rank +1 # 배낀 부분 끝
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data2)
df
```

```python
df.to_excel('1~200위.xlsx')
```
